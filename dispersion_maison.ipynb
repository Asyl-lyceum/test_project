```python
import pandas as pd
import matplotlib.pyplot as plt
bikes = pd.read_csv('bikes.csv', sep=';', encoding='latin1', parse_dates=['Date'], dayfirst=True, index_col='Date')
bikes
```

```python
maison_bikes = bikes[['Maisonneuve 1']].copy()
maison_bikes
```

```python
avg = maison_bikes['Maisonneuve 1'].sum()/maison_bikes['Maisonneuve 1'].count()
avg = round(avg,2)
avg
maison_bikes['average'] = avg
maison_bikes
```

```python
moda = maison_bikes['Maisonneuve 1'].value_counts().index[0]
maison_bikes['moda'] = moda
maison_bikes
```

```python
c = int((maison_bikes['Maisonneuve 1'].shape[0])/2)
c
```

```python
math = maison_bikes['Maisonneuve 1'] - avg
maison_bikes['math'] = round(math**2,2)
maison_bikes
```

```python
total = maison_bikes['math'].sum()
total = round(total)
total
```

```python
dispersion = total/maison_bikes['Maisonneuve 1'].count()
dispersion
```

```python
x_avg = maison_bikes['Maisonneuve 1'].mean()
x1 = maison_bikes['Maisonneuve 1']
((x1 - x_avg)**2).mean()
```

```python
def variance(data, street):
    x_avg = data[street].mean()
    x1 = data[street]
    variance = ((x1 - x_avg)**2).mean()
    return variance
print(variance(bikes, 'Maisonneuve 1'))
```
